```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
x_spdiff = np.arange(0, 256, 1)
x_acc = np.arange(0, 256, 1)
x_throttle  = np.arange(0, 256, 1)
```

```python
spdiff_NL = fuzz.trapmf(x_spdiff, [0, 0, 31, 63])
spdiff_NM = fuzz.trimf(x_spdiff, [31, 63, 95])
spdiff_NS = fuzz.trimf(x_spdiff, [63, 95, 127])
spdiff_ZE = fuzz.trimf(x_spdiff, [95, 127, 159])
spdiff_PS = fuzz.trimf(x_spdiff, [127, 159, 191])
spdiff_PM = fuzz.trimf(x_spdiff, [159, 191, 223])
spdiff_PL = fuzz.trapmf(x_spdiff, [191, 223, 255, 255])

acc_NL = fuzz.trapmf(x_acc, [0, 0, 31, 63])
acc_NM = fuzz.trimf(x_acc, [31, 63, 95])
acc_NS = fuzz.trimf(x_acc, [63, 95, 127])
acc_ZE = fuzz.trimf(x_acc, [95, 127, 159])
acc_PS = fuzz.trimf(x_acc, [127, 159, 191])
acc_PM = fuzz.trimf(x_acc, [159, 191, 223])
acc_PL = fuzz.trapmf(x_acc, [191, 223, 255, 255])

throttle_NL = fuzz.trapmf(x_throttle, [0, 0, 31, 63])
throttle_NM = fuzz.trimf(x_throttle, [31, 63, 95])
throttle_NS = fuzz.trimf(x_throttle, [63, 95, 127])
throttle_ZE = fuzz.trimf(x_throttle, [95, 127, 159])
throttle_PS = fuzz.trimf(x_throttle, [127, 159, 191])
throttle_PM = fuzz.trimf(x_throttle, [159, 191, 223])
throttle_PL = fuzz.trapmf(x_throttle, [191, 223, 255, 255])
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(8, 9))

ax0.plot(x_spdiff, spdiff_NL, 'b', linewidth=1.5, label='NL')
ax0.plot(x_spdiff, spdiff_NM, 'g', linewidth=1.5, label='NM')
ax0.plot(x_spdiff, spdiff_NS, 'r', linewidth=1.5, label='NS')
ax0.plot(x_spdiff, spdiff_ZE, 'y', linewidth=1.5, label='ZE')
ax0.plot(x_spdiff, spdiff_PS, 'r', linewidth=1.5, label='PS')
ax0.plot(x_spdiff, spdiff_PM, 'g', linewidth=1.5, label='PM')
ax0.plot(x_spdiff, spdiff_PL, 'b', linewidth=1.5, label='PL')
ax0.set_title('Speed Difference')
ax0.legend()

ax1.plot(x_acc, acc_NL, 'b', linewidth=1.5, label='NL')
ax1.plot(x_acc, acc_NM, 'g', linewidth=1.5, label='NM')
ax1.plot(x_acc, acc_NS, 'r', linewidth=1.5, label='NS')
ax1.plot(x_acc, acc_ZE, 'y', linewidth=1.5, label='ZE')
ax1.plot(x_acc, acc_PS, 'r', linewidth=1.5, label='PS')
ax1.plot(x_acc, acc_PM, 'g', linewidth=1.5, label='PM')
ax1.plot(x_acc, acc_PL, 'b', linewidth=1.5, label='PL')
ax1.set_title('Acceleration')
ax1.legend()

ax2.plot(x_throttle, throttle_NL, 'b', linewidth=1.5, label='NL')
ax2.plot(x_throttle, throttle_NM, 'g', linewidth=1.5, label='NM')
ax2.plot(x_throttle, throttle_NS, 'r', linewidth=1.5, label='NS')
ax2.plot(x_throttle, throttle_ZE, 'y', linewidth=1.5, label='ZE')
ax2.plot(x_throttle, throttle_PS, 'r', linewidth=1.5, label='PS')
ax2.plot(x_throttle, throttle_PM, 'g', linewidth=1.5, label='PM')
ax2.plot(x_throttle, throttle_PL, 'b', linewidth=1.5, label='PL')
ax2.set_title('Throttle')
ax2.legend()

# Turn off top/right axes
for ax in (ax0, ax1, ax2):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

```python
ip_spd=float(input("Enter speed difference as a crisp value "))
ip_acc=float(input("Enter acceleration as a crisp value "))
```

```python
#some zero some nonzero
spdiff_level_NL = fuzz.interp_membership(x_spdiff, spdiff_NL, ip_spd)
spdiff_level_NM = fuzz.interp_membership(x_spdiff, spdiff_NM, ip_spd)
spdiff_level_NS = fuzz.interp_membership(x_spdiff, spdiff_NS, ip_spd)
spdiff_level_ZE = fuzz.interp_membership(x_spdiff, spdiff_ZE, ip_spd)
spdiff_level_PS = fuzz.interp_membership(x_spdiff, spdiff_PS, ip_spd)
spdiff_level_PM = fuzz.interp_membership(x_spdiff, spdiff_PM, ip_spd)
spdiff_level_PL = fuzz.interp_membership(x_spdiff, spdiff_PL, ip_spd)

acc_level_NL = fuzz.interp_membership(x_acc, acc_NL, ip_acc)
acc_level_NM = fuzz.interp_membership(x_acc, acc_NM, ip_acc)
acc_level_NS = fuzz.interp_membership(x_acc, acc_NS, ip_acc)
acc_level_ZE = fuzz.interp_membership(x_acc, acc_ZE, ip_acc)
acc_level_PS = fuzz.interp_membership(x_acc, acc_PS, ip_acc)
acc_level_PM = fuzz.interp_membership(x_acc, acc_PM, ip_acc)
acc_level_PL = fuzz.interp_membership(x_acc, acc_PL, ip_acc)
```

```python
#Rule 1
active_rule1 = np.fmin(spdiff_level_NL, acc_level_ZE)
throttle_activation_PL1 = np.fmin(active_rule1, throttle_PL)
```

```python
#Rule 2
active_rule2 = np.fmin(spdiff_level_ZE, acc_level_NL)
throttle_activation_PL2 = np.fmin(active_rule2, throttle_PL)
```

```python
#Rule 3
active_rule3 = np.fmin(spdiff_level_NM, acc_level_ZE)
throttle_activation_PM1 = np.fmin(active_rule3, throttle_PM)
```

```python
#Rule 4
active_rule4 = np.fmin(spdiff_level_ZE, acc_level_NM)
throttle_activation_PM2 = np.fmin(active_rule4, throttle_PM)
```

```python
#Rule 5
active_rule5 = np.fmin(spdiff_level_NS, acc_level_ZE)
throttle_activation_PS1 = np.fmin(active_rule5, throttle_PS)
```

```python
#Rule 6
active_rule6 = np.fmin(spdiff_level_ZE, acc_level_NS)
throttle_activation_PS2 = np.fmin(active_rule6, throttle_PS)
```

```python
#Rule 7
active_rule7 = np.fmin(spdiff_level_NS, acc_level_PS)
throttle_activation_PS3 = np.fmin(active_rule7, throttle_PS)
```

```python
#Rule 8
active_rule8 = np.fmin(spdiff_level_ZE, acc_level_ZE)
throttle_activation_ZE = np.fmin(active_rule8, throttle_ZE)
```

```python
#Rule 9
active_rule9 = np.fmin(spdiff_level_PS, acc_level_NS)
throttle_activation_NS1 = np.fmin(active_rule9, throttle_NS)
```

```python
#Rule 10
active_rule10 = np.fmin(spdiff_level_ZE, acc_level_PS)
throttle_activation_NS2 = np.fmin(active_rule10, throttle_NS)
```

```python
#Rule 11
active_rule11 = np.fmin(spdiff_level_PS, acc_level_ZE)
throttle_activation_NS3 = np.fmin(active_rule11, throttle_NS)
```

```python
#Rule 12
active_rule12 = np.fmin(spdiff_level_ZE, acc_level_PM)
throttle_activation_NM1 = np.fmin(active_rule12, throttle_NM)
```

```python
#Rule 13
active_rule13 = np.fmin(spdiff_level_PM, acc_level_ZE)
throttle_activation_NM2 = np.fmin(active_rule13, throttle_NM)
```

```python
#Rule 14
active_rule14 = np.fmin(spdiff_level_ZE, acc_level_PL)
throttle_activation_NL1 = np.fmin(active_rule14, throttle_NL)
```

```python
#Rule 15
active_rule15 = np.fmin(spdiff_level_PL, acc_level_ZE)
throttle_activation_NL2 = np.fmin(active_rule15, throttle_NL)
```

```python
throttle_activation_PL=np.fmax(throttle_activation_PL1, throttle_activation_PL2)
throttle_activation_PM=np.fmax(throttle_activation_PM1, throttle_activation_PM2)
throttle_activation_PS=np.fmax(throttle_activation_PS3,np.fmax(throttle_activation_PS1, throttle_activation_PS2))
throttle_activation_NL=np.fmax(throttle_activation_NL1, throttle_activation_NL2)
throttle_activation_NM=np.fmax(throttle_activation_NM1, throttle_activation_NM2)
throttle_activation_NS=np.fmax(throttle_activation_NS3,np.fmax(throttle_activation_NS1, throttle_activation_NS2))
throttle_activation_P=np.fmax(throttle_activation_PL,np.fmax(throttle_activation_PM, throttle_activation_PS))
throttle_activation_N=np.fmax(throttle_activation_NL,np.fmax(throttle_activation_NM, throttle_activation_NS))
aggregate=np.fmax(throttle_activation_ZE,np.fmax(throttle_activation_N,throttle_activation_P))
```

```python
throttle0 = np.zeros_like(x_throttle)
# Visualization
fig, ax0 = plt.subplots(figsize=(8, 3))

ax0.plot(x_throttle, throttle_NL, 'y', linewidth=1, linestyle='--')
ax0.plot(x_throttle, throttle_NM, 'y', linewidth=1, linestyle='--')
ax0.plot(x_throttle, throttle_NS, 'y', linewidth=1, linestyle='--')
ax0.plot(x_throttle, throttle_ZE, 'y', linewidth=1, linestyle='--')
ax0.plot(x_throttle, throttle_PS, 'y', linewidth=1, linestyle='--')
ax0.plot(x_throttle, throttle_PM, 'y', linewidth=1, linestyle='--')
ax0.plot(x_throttle, throttle_PL, 'y', linewidth=1, linestyle='--')

ax0.fill_between(x_throttle, throttle0, aggregate, facecolor='Orange', alpha=0.7)
ax0.set_title('Aggregated Output membership')

# Turn off top/right axes
for ax in (ax0,):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

```python
#defuzzification
if (aggregate.sum()==0):
    print("Total area is zero in defuzzification!")
else:
    throttle_c = fuzz.defuzz(x_throttle, aggregate, 'centroid')
    print(throttle_c)
```

```python
#defuzzification
if (aggregate.sum()==0):
    print("Total area is zero in defuzzification!")
else:
    throttle_min = fuzz.defuzz(x_throttle, aggregate, 'som')
    print(throttle_min)
```

```python
#defuzzification
if (aggregate.sum()==0):
    print("Total area is zero in defuzzification!")
else:
    throttle_max = fuzz.defuzz(x_throttle, aggregate, 'lom')
    print(throttle_max)
```

```python
if (aggregate.sum()==0):
    print("Total area is zero in defuzzification!")
else:
    
    labels = ['centroid', 'bisector', 'mean of maximum', 'min of maximum',
          'max of maximum']
    xvals = [throttle_c,
         throttle_b,
         throttle_mean,
         throttle_min,
         throttle_max]
    colors = ['r', 'b', 'g', 'c', 'm']
    ymax = [fuzz.interp_membership(x_throttle, aggregate, i) for i in xvals]

    # Display and compare defuzzification results against membership function
    plt.figure(figsize=(8, 5))

    plt.plot(x_throttle, aggregate, 'k')
    for xv, y, label, color in zip(xvals, ymax, labels, colors):
        plt.vlines(xv, 0, y, label=label, color=color)
    plt.ylabel('Fuzzy membership')
    plt.xlabel('Universe variable (arb)')
    plt.ylim(-0.1, 1.1)
    plt.legend(loc=2)

    plt.show()
```
